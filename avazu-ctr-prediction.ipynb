```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test = train_test_split(train_data, test_size=.3, random_state=0)
```

```python
model_feature = 'click|banner_pos|device_conn_type|device_type|h|C1|C15|C16|C14|C17|C18|C19|C20|C21'
model_index = ['click', 'banner_pos', 'device_conn_type', 'device_type', 'h', 'C1', 'C15', 'C16',
               'C14', 'C17', 'C18', 'C19', 'C20', 'C21']
train_df = X_train.filter(regex=model_feature)
index = model_index
train_df = train_df.reindex(index, axis=1)
train_df.info()
```

```python
test_df = X_test.filter(regex=model_feature)
index = model_index
test_df = test_df.reindex(index, axis=1)
test_df.info()
```

```python
# LR
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
x_train = train_df.as_matrix()[:,1:]
y_train = train_df.as_matrix()[:,0]
y_train = y_train.astype('int')
clf = LogisticRegression(C=1.0, penalty='l1', tol=1e-3)
# clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(14, 5), random_state=0)
clf.fit(x_train, y_train)
x_test = test_df.as_matrix()[:,1:]
y_test = test_df.as_matrix()[:,0]
y_test = y_test.astype('int')
clf.score(x_test, y_test)
```

```python
import re
replace_reg = re.compile(r'\[|\]')
coef_feature = pd.DataFrame({'columns':list(train_df.columns)[1:],
                           'coef':[replace_reg.sub('',str(value)) for value in list(clf.coef_.T)]})
coef_b = pd.Series({'columns':'intercept_','coef':str(clf.intercept_[0])})
coef_feature = coef_feature.append(coef_b, ignore_index=True)
coef_feature
```

```python
cv_df = X_test.filter(regex=model_feature)
index=model_index
cv_df = cv_df.reindex(index, axis=1)
x_test = cv_df.as_matrix()[:, 1:]
y_test = cv_df.as_matrix()[:, 0]
x_test, y_test
```

```python
proba = clf.predict_proba(x_test)[:,1]
X_test['proba'] = proba
predictions = clf.predict(x_test)
X_test['predictions'] = predictions
```

```python
X_test.head(10)
```

```python
from sklearn.metrics import precision_recall_curve, roc_curve, auc
from sklearn.metrics import classification_report
y_test = y_test.astype(int)
precision, recall, threholds = precision_recall_curve(y_test, X_test['proba'])
precision, recall, threholds
report = X_test['proba'] > 0.5
y_test = [int(x) for x in y_test]
report_result = classification_report(y_test, report, target_names = ['neg', 'pos'])
report_result
```

```python
# MLP
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
x_train = train_df.as_matrix()[:,1:]
y_train = train_df.as_matrix()[:,0]
y_train = y_train.astype('int')
# clf = MLPClassifier()
clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 14), random_state=0)
clf.fit(x_train, y_train)
x_test = test_df.as_matrix()[:,1:]
y_test = test_df.as_matrix()[:,0]
y_test = y_test.astype('int')
clf.score(x_test, y_test)
```

```python
# XGB
import xgboost as xgb
x_train = train_df.as_matrix()[:,1:]
y_train = train_df.as_matrix()[:,0]
y_train = y_train.astype('int')
clf = xgb.XGBClassifier()
clf.fit(x_train, y_train)
x_test = test_df.as_matrix()[:,1:]
y_test = test_df.as_matrix()[:,0]
y_test = y_test.astype('int')
clf.score(x_test, y_test)
```

```python
# DNN
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.wrappers.scikit_learn import KerasClassifier
x_train = train_df.as_matrix()[:,1:]
y_train = train_df.as_matrix()[:,0]
y_train = y_train.astype('int')

model = Sequential([
    Dense(5, input_shape=(14,)),
    Activation('relu'),
    Dense(2),
    Activation('sigmoid'),
])
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])
clf = KerasClassifier(build_fn=model)

clf.fit(x_train, y_train)
x_test = test_df.as_matrix()[:,1:]
y_test = test_df.as_matrix()[:,0]
y_test = y_test.astype('int')
clf.score(x_test, y_test)
```
